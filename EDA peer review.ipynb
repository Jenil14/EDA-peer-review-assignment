```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
#Reading data frame
df = pd.read_csv('C:/Users/OWNER/Desktop/Machine Learning Coursera/EDA peer review assignment/Churn_Modelling.csv',delimiter=',')
df.shape
#hellohow are
```

```python
#Checking for missing values
df.isnull().sum()
# There are no missing values
```

```python
#Dropping irrelevant columns
df = df.drop(columns=['RowNumber', 'CustomerId', 'Surname'])
df.head()
```

```python
# Pie chart of customers exited and not exited
labels = 'Exited', 'Not Exited'
sizes = [df.Exited[df['Exited']==1].count(), df.Exited[df['Exited']==0].count()]
fig1, ax1 = plt.subplots(figsize=(10,8))
ax1.pie(sizes, labels= labels, autopct='%1.2f%%', textprops={'color':"white"})
ax1.axis('equal')
plt.title("Proportion of customer churned and retained", size = 20, color = 'white')
plt.show
```

```python
fig, axarr = plt.subplots(3, 2, figsize=(20, 12))
sns.countplot(x='Geography', hue = 'Exited',data = df, ax=axarr[0][0])
sns.countplot(x='Gender', hue = 'Exited',data = df, ax=axarr[0][1])
sns.countplot(x='HasCrCard', hue = 'Exited',data = df, ax=axarr[1][0])
sns.countplot(x='IsActiveMember', hue = 'Exited',data = df, ax=axarr[1][1])
sns.countplot(x='Age', hue='Exited',data=df, ax=axarr[2][0])
sns.countplot(x='Tenure', hue='Exited',data=df, ax=axarr[2][1])
```

According to the plots above, the majority of people who churned are from France and Germany, followed by Spain.
There are more females customers who exited the bank than male customers. 
Active members are inversely proportional to the people who churned the bank. 
According to the data, the ratio of customers with credit card who have churned the bank to the customers without credit card who have churned the bank is low. 
According to the data, tenure doesn't play important role with churn rate. However, from the data, we see that customers within the first year at the customers with 10 years at the bank had lower churn rate compared to customers within 1 to 9 years of tenure. However, customers with 1 to 9 years of tenure had almost similar churn rate. 


```python
fig, axarr = plt.subplots(3, 2, figsize=(20, 12))

sns.boxplot(y='Age',x = 'Exited', hue = 'Exited',data = df , ax=axarr[0][0])
sns.boxplot(y='Balance',x = 'Exited', hue = 'Exited',data = df, ax=axarr[0][1])
sns.boxplot(y='Tenure',x = 'Exited', hue = 'Exited',data = df, ax=axarr[1][0])
sns.boxplot(y='CreditScore',x = 'Exited', hue = 'Exited',data = df, ax=axarr[1][1])
sns.boxplot(y='NumOfProducts',x = 'Exited', hue = 'Exited',data = df, ax=axarr[2][0])
sns.boxplot(y='EstimatedSalary',x = 'Exited', hue = 'Exited',data = df, ax=axarr[2][1])
```

1. The older people are churning more than the younger ones. The bank can take an action wherein they give more attention to the older generation by bringing out some policies in favor of older people so that they retain.
2. Unfortunately, people with higher balance are churning. This might affect the banks giving loans to their customers. 
3. There isn't much effect on churn rate with average tenure people.
4. There is no relationship between the churn rate and CreditScore, NumberofProducts, EstimatedSalary.
